Machine learning
Machine Learning is a method of teaching computers to learn from data and improve their performance automatically. What is Machine Learning?
Machine Learning (ML) is a branch of Artificial Intelligence (AI) that enables computers to learn from data and improve their performance without being explicitly programmed.

Definition:

"Machine Learning is the ability of a computer system to learn from data and make predictions or decisions without being explicitly programmed."

Types of Machine Learning

Supervised Learning
In Supervised Learning, the model is trained using labeled data, where both input and output are known.

Features: Uses labeled data. Predicts outcomes based on learned patterns. Examples: Email Spam Detection House Price Prediction Student Performance Prediction Common Algorithms: Linear Regression Logistic Regression Decision Tree Random Forest Example:

A model learns the relationship between study hours and exam marks to predict future results.

Unsupervised Learning
In Unsupervised Learning, the model is trained using unlabeled data.

Features: No predefined output. Finds hidden patterns and relationships in data. Examples: Customer Segmentation Recommendation Systems Market Basket Analysis Common Algorithms: K-Means Clustering Hierarchical Clustering Apriori Algorithm Example:

An e-commerce company groups customers based on similar purchasing behavior.

semi supervised
Semi-Supervised Learning is a combination of Supervised Learning and Unsupervised Learning.

It uses:

A small amount of labeled data A large amount of unlabeled data

This approach is useful when labeling data is expensive or time-consuming.

Definition:

"Semi-Supervised Learning is a machine learning technique that trains a model using both labeled and unlabeled data."

Example:

Suppose you have:

1000 images of animals Only 100 images are labeled (cat, dog, etc.) 900 images are unlabeled

The model uses the labeled images to learn and the unlabeled images to improve its understanding and accuracy.

Applications: Image Classification Speech Recognition Medical Diagnosis Web Content Classification Advantages:

✅ Requires less labeled data ✅ More accurate than purely unsupervised learning ✅ Reduces the cost of data labeling

Example in Real Life:

A photo app may have a few labeled pictures of cats and dogs, while thousands of unlabeled photos are available. The model learns from both to improve its classification performance.

Reinforcement Learning
In Reinforcement Learning, an agent learns by interacting with its environment and receiving rewards or penalties.

Features: Learns through trial and error. Goal is to maximize rewards. Examples: Self-Driving Cars Robotics Game Playing (Chess, AlphaGo) Example:

A robot learns to navigate a room by receiving rewards for correct actions and penalties for mistakes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_bike_data.csv")
```

```python
df
```

```python
df.drop("Unnamed: 0",axis = 1,inplace=True)
```

```python
x=df.drop("price",axis = 1)
y=df[["price"]]
```

```python
x
```

```python
y
```

```python
#! pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x,y, random_state =42, test_size = 0.2)
```

```python
x_train.shape ,y_train.shape ,x_test.shape , y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
bike_1 = lr.predict([[1.0,	1.0	,17654.0,	1	,3.0,	110.0,	1]])
bike_1
```

```python
print(lr.score(x_train, y_train))
print(lr.score(x_test, y_test))
```
